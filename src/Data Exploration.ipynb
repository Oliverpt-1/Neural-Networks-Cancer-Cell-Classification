```python
#Data Exploration
    # Correlation heatmap
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

column_names = [
    "ID", "Diagnosis", 
    "Mean Radius", "Mean Texture", "Mean Perimeter", "Mean Area", "Mean Smoothness",
    "Mean Compactness", "Mean Concavity", "Mean Concave Points", "Mean Symmetry", "Mean Fractal Dimension",
    "Radius SE", "Texture SE", "Perimeter SE", "Area SE", "Smoothness SE",
    "Compactness SE", "Concavity SE", "Concave Points SE", "Symmetry SE", "Fractal Dimension SE",
    "Worst Radius", "Worst Texture", "Worst Perimeter", "Worst Area", "Worst Smoothness",
    "Worst Compactness", "Worst Concavity", "Worst Concave Points", "Worst Symmetry", "Worst Fractal Dimension"
]


# Load the data (assuming it's comma-separated)
file_path = "wdbc.data"  # Update with the correct path
df = pd.read_csv(file_path, header=None, names=column_names)

df['Diagnosis'] = df['Diagnosis'].map({'M': 1, 'B': 0})  # Convert M to 1, B to 0

df_numeric = df.drop(columns=['ID'], errors='ignore')

corr_matrix = df_numeric.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of WDBC Features")
plt.show()


plt.figure(figsize=(8,5))
sns.histplot(df["Mean Radius"], bins=30, kde=True, color="blue")
plt.xlabel("Mean Radius")
plt.ylabel("Frequency")
plt.title("Histogram of Mean Radius")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df["Worst Radius"], bins=30, kde=True, color="blue")
plt.xlabel("Worst Radius")
plt.ylabel("Frequency")
plt.title("Histogram of Worst Radius")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df["Worst Concave Points"], bins=30, kde=True, color="blue")
plt.xlabel("Worst Concave Points")
plt.ylabel("Frequency")
plt.title("Histogram of Worst Concave Points")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df["Worst Perimeter"], bins=30, kde=True, color="blue")
plt.xlabel("Worst Perimeter")
plt.ylabel("Frequency")
plt.title("Histogram of Worst Perimeter")
plt.show()
```
